```python
df = pd.read_csv("orders.csv")
df.head()
```

```python
df_orders = df.copy()
display(df_orders.info())
display(df_orders.describe().T)
```

```python
# Question 1 - 

# Deleting duplicates:

# 1. Finding duplicate items by item_name

dup_items = df_Items[df_Items['item_name'].duplicated(keep=False)].sort_values('item_name')
display(dup_items.head())
display(dup_items.tail())
```

```python
# 2. Count non-empty columns in each duplicate row

dup_items['non_nan_count'] = dup_items.notna().sum(axis=1)
dup_items = dup_items.sort_values(['item_name', 'non_nan_count'], ascending=[True, False])
display(dup_items.head())
display(dup_items.tail())
```

```python
# # 3. Choosing the best row from all duplicates

kept = dup_items.drop_duplicates(subset=['item_name'], keep='first')

removed = dup_items[~dup_items['id'].isin(kept['id'])]

display('Removed duplicate rows:', removed)
print()
display('Kept rows:', kept)
```

```python
# 4. Merge removed and kept rows based on item_name to track replacements

merged = removed.merge(kept, how='inner', on='item_name', suffixes=('_removed', '_kept'))

display(merged[['id_removed', 'id_kept']])
```

```python
# 5. Create a dictionary

dict_replace = dict(zip(merged['id_removed'], merged['id_kept']))
dict_replace
```

```python
# 6. Replace item_id in the order_item

df_order_item["item_id"] = df_order_item["item_id"].replace(dict_replace)
df_order_item
```

```python
# 7. Remove customers with 3 or more missing fields

df_customers['missing_count'] = df_customers.isna().sum(axis=1)
df_customers_clean = df_customers[df_customers['missing_count'] < 3]
display(df_customers[df_customers['missing_count'] >= 3])
```

```python
# 8. Remove order_item rows that belong to removed customers

valid_customer_ids = df_customers_clean['id']
valid_order_ids = df_orders[df_orders['customer_id'].isin(valid_customer_ids)]['order_id']
df_order_item = df_order_item[df_order_item['order_id'].isin(valid_order_ids)]
df_order_item
```

```python
# Question 2 -

# Information analysis:


# 3.

df_order_item_price = pd.merge(df_order_item, df_Items[['id', 'item_price']], left_on='item_id', right_on='id', how='left')

df_order_item_price['total_price'] = df_order_item_price['quantity'] * df_order_item_price['item_price']

display(df_order_item_price[['order_id', 'item_id', 'quantity', 'item_price', 'total_price']])
```

```python
# 4.

order_total = df_order_item_price.groupby('order_id')['total_price'].sum().reset_index()

display(order_total.rename(columns={'total_price': 'order_total_price'}))
```

```python
# 5.

most_expensive = order_total['total_price'].max()

cheapest = order_total['total_price'].min()

average = order_total['total_price'].mean()

display(f"Most expensive purchase: {most_expensive:.2f}")
print()
display(f"Cheapest purchase: {cheapest:.2f}")
print()
display(f"Average purchase amount: {average:.2f}")
```

```python
# Question 2 -

# graph -


# 3.

df_age_clean = df_customers.dropna(subset=['age'])

plt.figure(figsize=(10, 6))
sns.histplot(data=df_age_clean, x='age', bins=20, kde=True)

plt.title("Histogram of Customer Age Groups")
plt.xlabel("Age")
plt.ylabel("Number of Customers")
plt.tight_layout()
plt.show()
```

```python
# 4.

customers_by_year = df_customers.dropna(subset=['joining_date']).copy()

customers_by_year['joining_date'] = pd.to_datetime(customers_by_year['joining_date'], dayfirst=True, errors='coerce')

customers_by_year['join_year'] = customers_by_year['joining_date'].dt.year

yearly_counts = customers_by_year['join_year'].value_counts().sort_index()

plt.figure(figsize=(8, 5))
sns.barplot(x=yearly_counts.index, y=yearly_counts.values)
plt.title("Number of New Customers by Year")
plt.xlabel("Year")
plt.ylabel("Number of Customers")
plt.tight_layout()
plt.show()
```

```python
# 5.

orders_by_month = df_orders.dropna(subset=['order_date']).copy()

orders_by_month['order_date'] = pd.to_datetime(orders_by_month['order_date'], errors='coerce')
                                               
orders_by_month['order_month'] = orders_by_month['order_date'].dt.month

monthly_sales = orders_by_month['order_month'].value_counts().sort_index()

plt.figure(figsize=(8, 5))
sns.barplot(x=monthly_sales.index, y=monthly_sales.values)
plt.title("Number of Orders per Month")
plt.xlabel("Month")
plt.ylabel("Number of Orders")
plt.tight_layout()
plt.show()
```

```python
# 6.

orders_with_age = pd.merge(df_orders, df_customers[['id', 'age']], left_on='customer_id', right_on='id', how='left')

orders_with_age = orders_with_age.dropna(subset=['age'])

plt.figure(figsize=(10, 6))
sns.histplot(data=orders_with_age, x='age', bins=20, kde=True)
plt.title("Distribution of Orders by Customer Age")
plt.xlabel("Customer Age")
plt.ylabel("Number of Orders")
plt.tight_layout()
plt.show()
```
